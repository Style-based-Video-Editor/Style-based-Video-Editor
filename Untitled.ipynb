```python
import cv2
x1=0
x2=150
y1=0
y2=150
def cropVid(x1,x2,y1,y2,path):
    vidcap = cv2.VideoCapture(path)
    fourcc = cv2.VideoWriter_fourcc(*"MP4V")
    fps = vidcap.get(cv2.CAP_PROP_FPS)

    out = cv2.VideoWriter('out.mp4', fourcc, fps, (int(y2-y1),int(x2-x1)))
    success,image = vidcap.read()
    count = 0
    while success:
        print(image.shape)
        image=image[y1:y2,x1:x2]
        print(image.shape)
        out.write(image)
        success,image = vidcap.read()
    out.release()
```
